```python
import pandas as pd
import numpy as np
import os
import re

def process_all_files(dir_path, disease_excel_path, output_csv):
    disease_df = pd.read_excel(disease_excel_path)
    
    all_data = []
    
    for file in os.listdir(dir_path):
        if file.lower().endswith('.csv'):
            filepath = os.path.join(dir_path, file)
            
            # Extract patient number from filename and convert to int
            match = re.search(r'_(\d{4})_', file)
            patient_number = int(match.group(1)) if match else None
            
            freq_df = pd.read_csv(filepath, index_col=0)
            
            filename_lower = file.lower()
            prediction = -1
            if "healthy" in filename_lower:
                prediction = 6
            elif "patho" in filename_lower and patient_number is not None:
                matched = disease_df[disease_df['Patient'] == patient_number]
                if not matched.empty:
                    prediction = matched.iloc[0]['Score (type)']
            
            # For each column, append as row with prediction label
            for col in freq_df.columns:
                if freq_df[col].isna().any():
                    continue
                col_data = freq_df[col].values
                row_with_label = np.append(col_data, prediction)
                all_data.append(row_with_label)
    
    combined_df = pd.DataFrame(all_data)
    combined_df.to_csv(output_csv, index=False, header=False)
    print(f"Processed data saved to {output_csv}")

# Usage example (update with correct paths)
process_all_files(
    dir_path='data/Resultados',
    disease_excel_path='data/0list.xlsx',
    output_csv='data/workingdataset3.csv'
)
```
